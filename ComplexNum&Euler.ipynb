#**Complex Numbers**


```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
z=complex(4,3)
plt.plot(np.real(z),np.imag(z),'ro')
plt.axis('square')
plt.xlim(0,5)
plt.ylim(0,5)
plt.grid(True)
plt.xlabel('real')
plt.ylabel('imaginary')
plt.title('Complex Number')
```

```python
magnitude_z=np.sqrt(np.real(z)**2 + np.imag(z)**2) # also use np.abs(z)
print('The amplitude is',magnitude_z)
```

```python
angle_z= math.atan(np.imag(z)/np.real(z)) # also use np.angle(z)
print('phase is',angle_z)
```

Euler's Formula

```python
x = np.linspace(-3,3,num=50)

plt.plot(x,np.exp(x),label='y=e^x')
plt.axis([min(x),max(x),0,np.exp(x[-1])])
plt.grid(True)
plt.legend()
plt.xlabel('x')
plt.show()
```

```python
#arbitraty vector magnitude
m = 4; # magnitude
k = np.pi/3; # phase
compnum = m*np.exp( 1j*k );

mag = np.abs(compnum);
phs = np.angle(compnum);

plt.polar([phs,phs],[0,mag])
plt.show()
```
